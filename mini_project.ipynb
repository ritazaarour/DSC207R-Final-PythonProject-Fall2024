## 3 & 4. Defining Objectives & Presenting Findings
**1.  Does a higher Freedom score mean a country is happier and trust in their government more?**

**Choice:** I chose a bubble chart to represent this question so I can easily see if these categories are positively correlated, negatively correlated, or not correlated at all. I also think that the size of the bubbles makes the chart a bit easier to read and makes outliers more visible when comparing these categories. I also color coded the bubbles by region so that we can see if there is a trend among the columns.

```python
plt.figure(figsize=(16,10))
sns.scatterplot(data=happiness_by_country, x='Freedom', y='Happiness Score', size='Trust (Government Corruption)', hue='Region', alpha=0.5, sizes=(10, 500))
plt.gca().set_title('Happiness Score vs Freedom vs Trust in Government')
plt.legend(fontsize=8)
plt.show()
```

**Inference:** There is a positive correlation between Happiness Score and Freedom. With that, many of the countries that have a higher freedom score and happiness score have larger sized bubbles, indicating that they also have more trust in their government/less corruption. However, we can see that a higher freedom and happiness score does not always equal more trust in government, and some countries with a lower happiness score or a lower freedom score have slightly larger bubbles.

**2. Do Family, Life Expectancy, and Generosity have an influence on Happiness Score?**

**Choice:** I chose a heatmap to represent this question because it makes comparing multiple columns compact and easy to read. The compactness and color gradient makes it clear to see which variables are correlated or not. In this map, I let the lighter colors equal more correlation while the darker colors meant less to no correlation.

```python
happiness_filtered = happiness_by_country[['Happiness Score', 'Family', 'Health (Life Expectancy)', 'Generosity']]
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(happiness_filtered.corr(), annot=True, fmt='.3', linewidth=1, cmap='crest_r')
plt.gca().set_title('Correlation Between Happiness, Family, Health, and Generosity')
plt.show()
```

**Inference:** A correlation of 1 means that the variables are directly proportional, while a correlation of 0 means that the variables are not correlated. We can see that Life Expectancy is the most correlated with Happiness Score (0.765). Family falls close behind in correlation with Happiness Score(0.739), but Family and Life Expectancy are not as correlated (0.588). Generosity is the least correlated among all of the columns, each having a correlation less than 0.2.

 **3. What region makes up the majority of the top 30 happiest countries?**

**Choice:** I chose a pie chart to represent this question because it is a simple way to see the proportions of regions that make up the top 30 countries in the dataset and is easy to see if a particular region sticks out over the others

```python
# filtering the top 30 countries
top_30 = happiness_by_country.head(n=30)

happiest_top_30 = top_30['Region'].value_counts()
happiest_top_30
```

```python
happiest_top_30.plot(kind='pie', colormap='Pastel2', autopct='%1.1f%%')
plt.title('Percentage Distribution of the 30 Happiest Countries by Region')
plt.show()
```

**Inference:** The Western European region makes up the majority of the top 30 happiest countries, nearing almost half of the filtered dataset. Latin America and Caribbean make up a bit more than 1/4 of the chart, and the rest of the 5 regions all total to a little over the final 1/4 of the dataset.
